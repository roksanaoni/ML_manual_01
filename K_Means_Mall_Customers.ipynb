```
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import sklearn.metrics as sm
```

```
df = pd.read_csv('Mall_Customers.csv')
```

```
print("First 5 rows:\n", df.head())
```

```
x = df[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']]
```

```
print("\nSelected Features (x):\n", x.head())
```

```
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(x)
```

```
predictedY = kmeans.labels_
```

```
print("\nCluster Centers:\n", kmeans.cluster_centers_)
```

```
df['Cluster'] = predictedY
```

```
colors = np.array(['red', 'green', 'blue'])
red_patch = mpatches.Patch(color='red', label='Cluster 0')
green_patch = mpatches.Patch(color='green', label='Cluster 1')
blue_patch = mpatches.Patch(color='blue', label='Cluster 2')
```

```
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.scatter(df['Age'], df['Spending Score (1-100)'], c=colors[predictedY])
plt.title('Clusters: Age vs Spending Score')
plt.xlabel('Age')
plt.ylabel('Spending Score')
plt.legend(handles=[red_patch, green_patch, blue_patch])
```
